```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(mouse_drug_data_to_load)
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load)



# Combine the data into a single dataset
df = pd.merge(mouse_drug_df, clinical_trial_df, on="Mouse ID")

# Display the data table for preview
df
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
df_group = df[["Drug", "Timepoint", "Tumor Volume (mm3)"]]
df_group = df.groupby(["Drug", "Timepoint"])


# Convert to DataFrame
df_grouped = df_group.mean().drop(["Metastatic Sites"], axis=1)


# Preview DataFrame
df_grouped = df_grouped.reset_index()
df_grouped.head(50)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_df = df_group.sem()
# Convert to DataFrame
sem_df = sem_df.drop(["Metastatic Sites", "Mouse ID"], axis=1)
sem_df = sem_df.reset_index()
# Preview DataFrame
sem_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
df_grouped = df_grouped.pivot(index="Timepoint", columns="Drug")
# Preview that Reformatting worked
df_grouped
```

```python
# Generate the Plot (with Error Bars)
drug = [
    "Capomulin", "Ceftamin", "Infubinol", "Ketapril", "Naftisol",
    "Placebo", "Propriva", "Ramicane", "Stelasyn", "Zoniferol"
]
marker = ["o", "x", ".", "v", "^", "<", "|", "1", "d", "+"]

d = sem_df["Drug"]
t = "Tumor Volume (mm3)"

df_slim = df_grouped[t]
plt.xlabel("Timepoint")
plt.ylabel(t)
x = df_slim.index
plt.grid(True, axis="y")
plt.title("Tumor Response to Treatment")

for i in range(len(drug)):
    y = df_slim[drug[i]]
    yerrs = sem_df[d == drug[i]][t]
    
    plt.errorbar(
        x, y, yerr=yerrs, marker=marker[i],
        markersize=5, label=drug[i], linestyle="dashed",
    )
plt.legend(loc="best", ncol=2, shadow=True, fontsize="small")
    
# Save the Figure
plt.savefig("tumor_response.png", dpi=300)

# Show Figure



```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
# Convert to DataFrame
meta_df = df.groupby(["Drug", "Timepoint"]).mean()
meta_df = meta_df.drop(columns=t)
meta_df
# Preview DataFrame
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
meta_sem = df.groupby(["Drug", "Timepoint"]).sem()
# Preview DataFrame
meta_sem = meta_sem.drop(columns=["Mouse ID", "Tumor Volume (mm3)"])
meta_sem
```

```python
meta_df_reset = meta_df.reset_index()
meta_df_reset
```

```python
# Generate the Plot (with Error Bars)
meta_sem = meta_sem.reset_index()

plt.xlabel("Timepoint")
plt.ylabel("Met. Sites")
x = meta_df_slim.index
plt.grid(True, axis="y")
plt.title("Metastatic Spread During Treatment")

for i in range(len(drug)):
    y = meta_df_slim[drug[i]]
    yerrs = meta_sem[d == drug[i]][m]
    
    plt.errorbar(
        x, y, yerr=yerrs, marker=marker[i],
        markersize=5, label=drug[i], linestyle="dashed",
    )
plt.legend(loc="best", ncol=2, shadow=True, fontsize="small")
    
# Save the Figure
plt.savefig("meta_spread.png", dpi=300)

# Show Figure
plt.show()

```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
c = "Count"
count_df = df.groupby(["Drug", "Timepoint"]).count()
# Convert to DataFrame
count_df = count_df.rename(columns={"Mouse ID": c}).drop(columns=[t, m]).reset_index()
# Preview DataFrame
count_df



```

```python
# Minor Data Munging to Re-Format the Data Frames
count_df_slim = count_df.pivot(index="Timepoint", columns="Drug")
count_df_slim = count_df_slim[c]
# Preview the Data Frame
count_df_slim
```

```python
# Generate the Plot (Accounting for percentages)
plt.xlabel("Timepoint")
plt.ylabel("Survival Rate (%)")
x = meta_df_slim.index
plt.grid(True)
plt.title("Mouse Survival Rate")
for i in range(len(drug)):
    y = 100 * count_df_slim[drug[i]]/count_df_slim[drug[i]].max()
    plt.plot(
        x, y, marker=marker[i],
        markersize=5, label=drug[i], linestyle="dashed",
    )
plt.legend(loc="best", ncol=2, shadow=True, fontsize="small")
    
# Save the Figure
plt.savefig("survival_rate.png", dpi=300)
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
perc_change = []
for i in range(len(drug)):
    perc_calc = 100 * (df_slim[drug[i]].values[len(df_slim[drug[i]]) - 1] - df_slim[drug[i]].values[0]) / df_slim[drug[i]].values[0]
    perc_change.append(perc_calc)

perc_drug = pd.DataFrame(zip(drug, perc_change), columns=["Drug", "Perc. Change"])

# Display the data to confirm
perc_drug
```

```python
drug_perc_tup = [(perc_drug["Drug"][i], perc_drug["Perc. Change"][i]) for i in (range(len(perc_drug)))]
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
for i in range(len(drug)):
    facecolor = ""
    edgecolor = ""
    if (drug_perc_tup[i][1] <= 0):
        facecolor = "green"
        edgecolor = "lightgreen"

    else:
        facecolor = "red"
        edgecolor = "salmon"

    ax.bar(drug_perc_tup[i][0], drug_perc_tup[i][1],
            facecolor=facecolor, edgecolor=edgecolor, width=1,
           )
    
# Orient widths. Add labels, tick marks, etc. 
ax.grid(True, axis="y")
plt.xticks(rotation=60)
plt.xlabel("Drug")
plt.ylabel("Tumor Volume Change (%)")
plt.title("Tumor Change Over 45 Day Treatment")
ax.axhline(c='k')
# y = -4.5, 1.5
# x = -.16, 8.8
# magic num = .997
for i in range(10):
    if (i == 0 or i == 7):
        ax.text(-.45 + i, -4.5, str(round(drug_perc_tup[i][1], 1)) + "%",
                color="white", weight="heavy", fontsize=7.5)
    else:
        ax.text(-.45 + i, 1.5, str(round(drug_perc_tup[i][1], 1)) + "%",
                color="white", weight="heavy", fontsize=7.5)


# Use functions to label the percentages of changes


# Call functions to implement the function calls


# Save the Figure
plt.tight_layout()
plt.savefig("tumor_perc_change.png", dpi=300)

# Show the Figure
plt.show()
```
